```python
import pandas as pd
import numpy as np

path = 'E:/Heon/2019.1학기/화철/input/'
col_dic = {'id' : 5,  'name' : 15, 'adress' : 50, 'car' : 15, 'work' : 15, 'phone' : 15}
col_dic2 = {'id' : 5,  'name' : 15, 'adress' : 50, 'car' : 15, 'work' : 15, 'phone' : 15, 'code' : 1}       

def get_old_master_data(people, col_dic) : 
        with open(path + 'old_master.txt', 'wb') as output:
            for person in people :                 
                i = 0
                for col in col_dic : 
                    info = bytearray(person[i], encoding='cp949')
                    if(len(info)>col_dic[col]) :
                        print("error")
                    while(len(info)<col_dic[col]) : 
                        info = info + b'\x00'
                    output.write(info)
                    i = i+1
            output.close()
            
def get_transaction_data(people, col_dic) : 
        with open(path + 'transaction.txt', 'wb') as output:
            for person in people :                 
                i = 0
                for col in col_dic : 
                    info = bytearray(person[i], encoding='cp949')
                    if(len(info)>col_dic[col]) :
                        print("error")
                    while(len(info)<col_dic[col]) : 
                        info = info + b'\x00'
                    output.write(info)
                    i = i+1
            output.close()
```

```python
transKey = transaction.read(5)
masterKey = old_master.read(5)
```

```python
with open(path + 'new_master.txt', 'wb') as new_master : 
    while(not(masterKey ==b'') and  not(transKey == b'') ) :
        print("트랜스 키 : " + str(transKey))
        print("마스터 키 : " + str(masterKey))
        print()
        if(int(masterKey) < int(transKey)) :
            #구 마스터 레코드를 신 마스터 화일에 복사하여라
            features = old_master.read(110)
            new_master.write(masterKey)
            new_master.write(features)
            #다음 구 마스터 레코드를 읽어라
            masterKey = old_master.read(5)
            

        elif(int(transKey)== int(masterKey)) : 
            features = transaction.read(110)
            code = transaction.read(1)
            
            #트랜잭션 레코드 형식은 어떠한가?
            if(code == b'1') : # 차 교환
                new_master.write(transKey)
                new_master.write(features)
                
                transKey = transaction.read(5) #다음 트렌젝션 레코드 읽어라
                old_master.read(110)
                masterKey = old_master.read(5) #다음 구 마스터 레코드 읽어라

            elif(code == b'3') : # 주소변경
                new_master.write(transKey)
                new_master.write(features)
                transKey = transaction.read(5)
                old_master.read(110)
                masterKey = old_master.read(5)


            elif(code == b'4') : # 전화번호 변경
                new_master.write(transKey)
                new_master.write(features)
                transKey = transaction.read(5)
                old_master.read(110)
                masterKey = old_master.read(5)

            elif(code == b'5') : # 직장 변경
                new_master.write(transKey)
                new_master.write(features)
                transKey = transaction.read(5)
                old_master.read(110)
                masterKey = old_master.read(5)
                

            elif(code == b'2') : # 제거될 레코드
                transKey = transaction.read(5) #다음 트렌젝션 레코드 읽어라
                old_master.read(110)
                masterKey = old_master.read(5) #다음 구 마스터 레코드 읽어라


            elif(code == b'0') : # 삽입할 레코드
                #다음 트렌젝션 레코드 읽어라
                transKey = transaction.read(5)
        
        else :
            features = transaction.read(110)
            code = transaction.read(1)
            #트렌젝션 레코드 형식은 어떠한가?
            
            if(code == b'0') : # 삽입할 레코드
                #트렌잭션 레코드를 신 마스터 화일에 복사하라
                new_master.write(transKey)
                new_master.write(features)
                #다음 트렌젝션 레코드 읽어라
                transKey = transaction.read(5)
            else : 
                #오류로서 트랜잭션 거절
                print("error in transKey  " + str(transKey) + "     code " +str(code))
                print()
                transKey = transaction.read(5)
        
        #트랜잭션 파일의 끝인가?
        if(transKey == b'') : #YES
            if(masterKey == b'') :
                new_master.close()
            else :
                restfile = old_master.read()
                new_master.write(restfile)
        
        else : #NO
            #마스터 화일의 끝인가?
            if(masterKey == b'') :  #YES
                features = transaction.read(110)
                code = transaction.read(1)
                
                if(code == b'1' and code == b'3' and code == b'4' and code == b'5' and code == b'2') : #트렌젝션 레코드에서 삭제나 수정이면 오류로써 거절
                    print("error")
                    transKey = transaction.read(5) #다음 트렌젝션 레코드 읽어라

                elif(code == b'0') : #삽입이면 신 마스터 화일에 복사.
                    new_master.write(transKey)
                    new_master.write(features)
                    transKey = transaction.read(5) 
                    
    new_master.close()
```
